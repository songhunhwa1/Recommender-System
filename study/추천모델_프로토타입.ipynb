```python
df1 = df[['Mem_No', 'name_sku']]
df1.columns = ['mem_no', 'name_sku']
df1['mem_no'] = df1['mem_no'].astype(str)
```

```python
df2 = df1.groupby(["mem_no", "name_sku"]).size().to_frame().reset_index()
df2.columns = ['mem_no', 'name_sku', 'ord_cnt']
```

```python
from surprise.model_selection import train_test_split, cross_validate, KFold, GridSearchCV
from surprise import Reader, Dataset, SVD, evaluate, accuracy
import surprise

reader = Reader()

data = Dataset.load_from_df(df2, reader)
```

```python
#trainset, testset = train_test_split(data, test_size=.15)

svdpp_options = {'n_epochs': 5, 'lr_all': 0.005, 'reg_all': 0.4}

# lr_all – The learning rate for all parameters.
# reg_all – The regularization term for all parameters.

svdpp = surprise.SVDpp() 
```

```python
kf = KFold(n_splits=5)

for trainset, testset in kf.split(data):
    
    svdpp.fit(trainset)
    pred = svdpp.test(testset)

    print(accuracy.rmse(pred, verbose=True))
```
